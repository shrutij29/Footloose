```python
directory = '/Users/Shruti/Downloads/Dance_Dataset/densepose/txt'

#get directories without hidden files
directories = [d for d in os.listdir(directory) if os.path.isdir(os.path.join(directory, d))]
directories.sort()

#Setup dictionary to collect file names
all_file_names = {}

#Identify all file names
for d in directories:
    #get file names without hidden files
    files = [f for f in os.listdir(os.path.join(directory, d)) if os.path.isfile(os.path.join(directory, d, f))]
    files.sort()
    all_file_names[d] = files
```

```python
#Count the number of videos in each video type
#Determined by checking last number string on each image and
#checking if it matches '0001'
num_videos = {}
first_img_indexes = {}

for d in directories:
    video_count = 0
    num_images = len(all_file_names[d])
    for i in range(num_images):
        if all_file_names[d][i].split('.')[0].split('_')[-1] == '0001':
            video_count += 1
    num_videos[d] = video_count
```

```python
#Print out summary results
print(f"{'dance':<9} | {'Num Images':<10} | {'Num Videos':<9}")
print("-"*35)
for k,v in all_file_names.items():
    print(f'{k:<9} | {len(v):<10} | {num_videos[k]}')
```

```python
average_images_per_video = {}

for d in directories:
    video_count = 0
    total_images = 0
    num_images = len(all_file_names[d])
    
    # Count the number of images per video in the directory
    for i in range(num_images):
        if all_file_names[d][i].split('.')[0].split('_')[-1] == '0001':
            video_count += 1
            if video_count > 1:
                # Calculate the average number of images per video
                average_images_per_video[d] = total_images / (video_count - 1)
                total_images = 0
        total_images += 1
        
    
    # Calculate the average number of images per video for the last video
    if video_count > 0:
        average_images_per_video[d] = total_images / video_count
```

```python
# Calculate average images per video and add to summary
print(f"{'dance':<9} | {'Num Images':<10} | {'Num Videos':<9} | {'Avg Images per Video':<20}")
print("-" * 60)
for k, v in all_file_names.items():
    num_images = len(v)
    num_videos_in_dir = num_videos[k]  # Get the number of videos in the directory
    avg_images_per_video = num_images / num_videos_in_dir if num_videos_in_dir > 0 else 0
    print(f'{k:<9} | {num_images:<10} | {num_videos_in_dir:<9} | {avg_images_per_video:<20.2f}')
```

```python
'''
Function: Number of people in a frame
'''
def person_count(file_name):
    '''Input a JSON file and get the number of people in each frame. '''
    f = open(file_name, "r")
    j = json.load(f)
    f.close

    people_count = len(j)

    return people_count
```

```python
# Setup lists to collect file paths
selected_files_ballet = []

# Identify all file names
for d in directories:
    if d == 'ballet':
        # Get file names without hidden files
        files = [f for f in os.listdir(os.path.join(directory, d)) if os.path.isfile(os.path.join(directory, d, f))]

        files.sort()
        for file_name in files:
            file_path = os.path.join(directory, d, file_name)
            # print(file_path)
            # Count people in each frame
            num_people = person_count(file_path)
            if num_people == 1:  # Check if there is only one person
                selected_files_ballet.append(file_path)

# Print the sizes of the selected files lists
print("Number of selected frames for ballet:", len(selected_files_ballet))
```

```python
# Directory containing ballet files
ballet_directory = '/Users/Shruti/Downloads/Dance_Dataset/densepose/txt/ballet'

# Initialize count
ballet_video_count = 0
ballet_video_one_person_frame_count = 0
ballet_video_one_person_frames = {}

# Iterate through files in the directory
for file_name in os.listdir(ballet_directory):

    file_path = os.path.join(ballet_directory, file_name)
    if person_count(file_path) == 1:
        ballet_video_one_person_frame_count+=1
            
    # Check if the file is a JSON file and represents the start of a video
    if file_name.endswith('.json') and file_name.split('_')[-1].split('.')[0] == '0001':
        ballet_video_count += 1
        ballet_video_one_person_frames[ballet_video_count] = ballet_video_one_person_frame_count
        ballet_video_one_person_frame_count = 0

ballet_video_one_person_frames_df = pd.DataFrame(ballet_video_one_person_frames.items(), columns=['Video', 'Frames with One Person'])

print("Number of ballet videos:", ballet_video_count)
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
ballet_video_one_person_frames_df
```

```python
# Sum the values in the dictionary
total_frames = sum(ballet_video_one_person_frames.values())

# Print the total
print("Total frames with only one person:", total_frames)
```
